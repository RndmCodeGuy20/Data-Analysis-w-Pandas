```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]    #- X-axis data
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]    #- Y-axis Data

plt.plot(years,yield_apples)
plt.xlabel("Year ➡️")
plt.ylabel("Yield (in tons per hectare) ➡️")
plt.show()
```

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]

plt.plot(years, apples)
plt.plot(years,oranges)

plt.xlabel("Years ➡️")
plt.ylabel("Yield (in tone per hectare) ➡️")
plt.title("Crop Yield in Kanto")
plt.legend(['Apples','Oranges'])

plt.show()
```

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]

plt.figure(figsize=(20,8))

plt.plot(years, apples,color='red',marker='s',alpha=0.5,mec='red',ls='-.',animated=1)
plt.plot(years,oranges,color='orange',marker='o',alpha=0.5,ls='-',mec='orange')

plt.xlabel("Years ➡️")
plt.ylabel("Yield (in tone per hectare) ➡️")
plt.title("Crop Yield in Kanto")
plt.legend(['Apples','Oranges'])

plt.show()
```
